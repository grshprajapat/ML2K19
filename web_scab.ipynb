```
from  bs4 import BeautifulSoup
```

```
import requests
```

```
url="https://en.wikipedia.org/wiki/Coprinellus_micaceus"
```

```
response=requests.get(url) # we are sending request to the url
response
```

```
response.text
```

```
#parsing library
soup=BeautifulSoup(response.text, "lxml")
```

```
soup
```

```
for Alink in soup.find_all('a'):
  print(Alink)
```

```
for ptag in soup.find_all('p'):
  print(ptag)
```

```
#now second loop for b tag
for btag in ptag.find_all('b'):
  print(btag.text)
```
